```python
import pandas as pd 
```

```python
import geopandas as gpd 
```

```python
recent_df = gpd.read_file("EmployeePayroll.csv")
```

# Reading the Data

```python
recent_df.head(10)
```

# Chaging the Data from Object to a Number

```python
recent_df.info()
```

```python
recent_df['Overtime Pay'] = recent_df['Overtime Pay'][recent_df["Overtime Pay"] != ""].astype(float)
```

```python
recent_df.describe().round()
```

# Subset of the Data 

```python
recent_df[["Department Title", "Overtime Pay"]].head(20)
```

# Group by Department and Sum the Overtime pay 

```python
department = recent_df.groupby(["Department Title"]).sum({"Overtime Pay"}).reset_index()
```

```python
department
```

# Group by year and Sum the Overtime

```python
year = recent_df.groupby(["Year"]).sum({"Overtime Pay"}).reset_index()
```

```python
year
```

# Percentage Change 

```python
year.loc[year["Year"] == "2013"]
```

```python
year.loc[year["Year"] == "2013"]["Overtime Pay"].values[0]
```

```python
#pct = change / original
original = year.loc[year["Year"] == "2013"]["Overtime Pay"].values[0]
new = year.loc[year["Year"] == "2018"]["Overtime Pay"].values[0]
change = new - original
print(change / original)
```
